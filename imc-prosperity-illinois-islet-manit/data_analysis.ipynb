```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_prices_round_1_day_1 = pd.read_csv("./island-data-bottle-round-1/prices_round_1_day_-1.csv", sep=';')
df_prices_round_1_day_1
```

```python
df_prices_round_1_day_1_pearls = df_prices_round_1_day_1[df_prices_round_1_day_1['product'] == 'PEARLS']
df_prices_round_1_day_1_bananas = df_prices_round_1_day_1[df_prices_round_1_day_1['product'] == 'BANANAS']
```

```python
plt.plot(df_prices_round_1_day_1_pearls['timestamp'], df_prices_round_1_day_1_pearls['mid_price'])
plt.show()
plt.clf()
```

```python
plt.plot(df_prices_round_1_day_1_bananas['timestamp'], df_prices_round_1_day_1_bananas['mid_price'])
plt.show()
plt.clf()
```

```python
movAvg = []
interval = 100

for i in range(interval,len(df_prices_round_1_day_1_bananas)):
    movAvg.append(sum(df_prices_round_1_day_1_bananas['mid_price'][i-interval:i])/interval)
plt.figure(figsize=(20,12))
plt.plot(df_prices_round_1_day_1_bananas['timestamp'][interval:], movAvg)
plt.plot(df_prices_round_1_day_1_bananas['timestamp'], df_prices_round_1_day_1_bananas['mid_price'])
plt.show()
plt.clf()

# momentum = []
# for i in range(interval,len(df_prices_round_1_day_1_bananas)):
#     m = (df_prices_round_1_day_1_bananas['mid_price'][i])/(df_prices_round_1_day_1_bananas['mid_price'][i-interval])
#     momentum.append(m)
# plt.figure(figsize=(20,12))
# plt.plot(df_prices_round_1_day_1_bananas['timestamp'][interval:], momentum)
# plt.show()
# plt.clf()
```

```python
df_prices_round_3_day_0 = pd.read_csv("./island-data-bottle-round-3/prices_round_3_day_0.csv", sep=';')
df_prices_round_3_day_0_gear = df_prices_round_3_day_0[df_prices_round_3_day_0['product'] == 'DIVING_GEAR']
df_prices_round_3_day_0_sightings = df_prices_round_3_day_0[df_prices_round_3_day_0['product'] == 'DOLPHIN_SIGHTINGS']
fig, ax1 = plt.subplots()
color = 'tab:red'
ax1.plot(df_prices_round_3_day_0_gear['timestamp'], df_prices_round_3_day_0_gear['mid_price'], color = color)
ax1.tick_params(axis='y', labelcolor=color)
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.plot(df_prices_round_3_day_0_sightings['timestamp'], df_prices_round_3_day_0_sightings['mid_price'])
ax2.tick_params(axis='y', labelcolor=color)
fig.tight_layout() 
fig.figure(figsize=(20,12))
plt.show()
plt.clf()
```

```python
df_prices_round_3_day_1 = pd.read_csv("./island-data-bottle-round-3/prices_round_3_day_1.csv", sep=';')
df_prices_round_3_day_1_gear = df_prices_round_3_day_1[df_prices_round_3_day_1['product'] == 'DIVING_GEAR']
df_prices_round_3_day_1_sightings = df_prices_round_3_day_1[df_prices_round_3_day_1['product'] == 'DOLPHIN_SIGHTINGS']

fig, ax1 = plt.subplots(figsize = (20,12))
color = 'tab:red'
ax1.plot(df_prices_round_3_day_1_gear['timestamp'], df_prices_round_3_day_1_gear['mid_price'], color = color)
ax1.tick_params(axis='y', labelcolor=color)
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.plot(df_prices_round_3_day_0_sightings['timestamp'] - 70000, df_prices_round_3_day_0_sightings['mid_price'], color = color)
ax2.tick_params(axis='y', labelcolor=color)
fig.tight_layout()
plt.show()
plt.clf()
```
